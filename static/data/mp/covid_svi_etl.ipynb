```python
# dependencies
import pandas as pd
```

# SVI Data Pre-Processing

```python
# read in SVI data
svi_df = pd.read_csv('SVI2020_US_COUNTY.csv', dtype={'ST': str, 'FIPS': str})
svi_df.tail()
```

```python
# create a subset dataframe with selected columns
svi_df_subset = svi_df[['ST','STATE','ST_ABBR','COUNTY','FIPS','LOCATION','E_TOTPOP','E_POV150','E_UNEMP','E_HBURD','E_NOHSDP','E_UNINSUR','E_AGE65','E_DISABL','E_SNGPNT','E_LIMENG']]
svi_df_subset.tail()
```

```python
svi_df_subset.info()
```

```python
# drop rows with values of 0
svi_df_filtered = svi_df_subset.loc[(svi_df_subset !=0 ).all(axis=1)]

# describe dataframe
svi_df_filtered.describe()
```

```python
# Calculate percentage of population affected per selected risk indicator
svi_df_filtered[['E_POV150_PER', 'E_UNEMP_PER','E_HBURD_PER','E_NOHSDP_PER','E_UNINSUR_PER','E_AGE65_PER','E_DISABL_PER','E_SNGPNT_PER','E_LIMENG_PER']] = svi_df_subset[['E_POV150','E_UNEMP','E_HBURD','E_NOHSDP','E_UNINSUR','E_AGE65','E_DISABL','E_SNGPNT','E_LIMENG']]\
    .apply(lambda x: x/svi_df_filtered['E_TOTPOP']*100).round(2)

# display data
svi_df_filtered.head()
```

```python
# drop columns
svi_data_final = svi_df_filtered.drop(columns=['E_POV150','E_UNEMP','E_HBURD','E_NOHSDP','E_UNINSUR','E_AGE65','E_DISABL','E_SNGPNT','E_LIMENG'])

# display data
svi_data_final
```

# County COVID Case Data Pre-Processing

```python
# read in county case data
county_df = pd.read_csv('covid_county.csv', dtype={'county_fips_code': str})
county_df.tail()
```

```python
# rename FIPS column
county_df = county_df.rename(columns={'county_fips_code': 'FIPS', 'COUNT': 'CASES'})
county_df
```

```python
# group case counts by county FIPS code
county_grouped_df = county_df.groupby('FIPS').agg({'CASES': sum})
county_grouped_df
```

# Merge COVID Case and SVI Data

```python
# merge datasets
merged_df = pd.merge(svi_data_final, county_grouped_df, how='inner', on='FIPS')
merged_df
```

```python
# calculate cases per capita
merged_df['CASES_PER_CAPITA'] = merged_df['CASES']/merged_df['E_TOTPOP']
merged_df.describe()
```

```python
# assign counties to quartiles of cases per capita
merged_df['QUARTILE'] = pd.qcut(merged_df['CASES_PER_CAPITA'], 4, labels=[1, 2, 3, 4])

merged_df
```

```python
# save to csv
merged_df.to_csv('covid_svi_county.csv')
```
